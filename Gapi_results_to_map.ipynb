```python
import folium
import numpy as np
import pandas as pd
import os
from folium import plugins #for heatmaps
```

```python
#Clean data before importing to the map
df = pd.read_csv('Results.csv')

df_colums = df[['name', 'rating', 'geometry.location.lat', 'geometry.location.lng' ]]

df_clean = df_colums.rename(columns={"geometry.location.lat": "lat", "geometry.location.lng": "lng"}, inplace=False)
```

```python
gmap = folium.Map(location=[54.679673, 25.283782], zoom_start=14) # Change coordinates or location for your desired location

for lat, lng, name, rating in zip(df_clean.lat, df_clean.lng, df_clean.name, df_clean.rating):
     folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='blue',
        fill = True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(gmap)

#Optional, just in case heatmaps is needed
hm_data = df_clean[["lat", "lng"]].to_numpy().tolist()
gmap.add_child(plugins.HeatMap(hm_data)) 

gmap
```
